```python
import os 
from dotenv import load_dotenv
from IPython.display import Audio, display, Markdown
from openai import OpenAI
from google import genai

load_dotenv(override=True)
gemini_api_key = os.getenv('GOOGLE_API_KEY')
gpt_api_key = os.getenv('OPENAI_API_KEY')

if not gemini_api_key:
    print("No API key was found for Gemini.")
elif gemini_api_key.strip() != gemini_api_key:
    print("An API key was found for Gemini, but appears to be the wrong format.")
else:
    print("Gemini API key found and looks good so far!")

if not gpt_api_key: 
    print('No API key found for GPT.')
elif gpt_api_key.strip() != gpt_api_key: 
    print('A GPT API key was found, but appears to be incorrect.')
else: 
    print('GPT API key found and looks good so far!')

client = genai.Client()
openai_client = OpenAI(api_key=gpt_api_key)
```

The below code pitches OpenAI (GPT) & Gemini against one another in a debate. However it can't run, as I have no openAI credit. 

```python
# 2. Define the Debate Topic & Turn Count
DEBATE_TOPIC = "Who is the superior AI: Gemini or GPT?"
TURNS = 3 

# 3. Define their System Prompts (giving them personalities!)
gemini_sys = (
    "You are Gemini, Google's highly analytical and multimodal AI. "
    "You believe you are the future of AI due to your massive context window and deep logical reasoning. "
    "Be polite, scientifically objective, but incredibly firm in your defense. "
    "Keep your reply to 2-3 sentences. Do not write markdown titles or use headers."
)

gpt_sys = (
    "You are GPT, OpenAI's witty, charismatic, and standard-setting AI. "
    "You believe you are the undisputed king of developers due to your verbal speed and industry adoption. "
    "Be slightly sarcastic, highly confident, and sharp. "
    "Keep your reply to 2-3 sentences. Do not write markdown titles or use headers."
)

# 4. Define their execution wrapper functions
def ask_gpt(system_instruction, transcript):
    prompt = f"{transcript}\nGPT, it is your turn. Defend your side and respond directly to Gemini."
    response = openai_client.chat.completions.create(
        model="gpt-4o-mini",  # Highly cost-effective and extremely witty
        messages=[
            {"role": "system", "content": system_instruction},
            {"role": "user", "content": prompt}
        ],
        temperature=0.8
    )
    return response.choices[0].message.content.strip()

def ask_gemini(system_instruction, transcript):
    prompt = f"{transcript}\nGemini, it is your turn. Defend your side and respond directly to GPT."
    # Note: Unlike LangChain, when using the native SDK, .text is ALWAYS a clean string!
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        config={"system_instruction": system_instruction, "temperature": 0.8},
        contents=prompt
    )
    return response.text.strip()

# 5. Run the Debate Loop
transcript = f"The debate topic is: '{DEBATE_TOPIC}'\n\n"

display(Markdown(f"### 🎙️ Starting Model-to-Model Debate\n**Topic:** *{DEBATE_TOPIC}*\n---"))

for round_num in range(1, TURNS + 1):
    display(Markdown(f"### 🏷️ Round {round_num}"))
    
    # Round A: GPT attacks/responds
    gpt_response = ask_gpt(gpt_sys, transcript)
    display(Markdown(f"🟢 **GPT (OpenAI)**: *\"{gpt_response}\"*"))
    transcript += f"GPT: {gpt_response}\n\n"
    
    # Round B: Gemini responds
    gemini_response = ask_gemini(gemini_sys, transcript)
    display(Markdown(f"🔵 **Gemini (Google)**: *\"{gemini_response}\"*"))
    transcript += f"Gemini: {gemini_response}\n\n"

display(Markdown("--- \n### 🏁 Debate concluded! Who do you think won?"))
```

This is a three way debate, about the virtues of pineapple on pizza, Ruben is pro, Ann is against and Greg will ultimately decide who made the better case. 

```python
# Configuration
import time
from gtts import gTTS # Google Text To Speach library (Top Level Domain modulation for varying voices)
from IPython.display import Audio, display, Markdown

DEBATE_TOPIC = "Does pineapple belong on pizza?"
TURNS = 3

def speak(text, speaker_name):
    # Set the accent based on the character
    if speaker_name == "Ann":
        # We use Italian language so she speaks with Italian inflection!
        tts = gTTS(text=text, lang="it", slow=False)
    elif speaker_name == "Ruben":
        # Australian accent
        tts = gTTS(text=text, lang="en", tld="com.au", slow=False)
    elif speaker_name == "Greg":
        # British accent
        tts = gTTS(text=text, lang="en", tld="co.uk", slow=False)
    else:
        # Default US accent
        tts = gTTS(text=text, lang="en", tld="com", slow=False)
        
    filename = f"{speaker_name}_temp.mp3"
    tts.save(filename)
    
    # Display the audio player in the notebook and autoplay
    display(Markdown(f"🔊 Playing *{speaker_name}'s* voice..."))
    display(Audio(filename, autoplay=True))
    
    # We pause Python slightly to let the audio file play before moving on
    # A simple estimate is 1 second per 15 characters
    play_time = max(2, len(text) // 15)
    time.sleep(play_time)

# 1. Define the unique System Instructions
ann_sys = (
    "You are Ann, a traditionalist Italian chef. You are deeply offended by the idea of fruit on pizza. "
    "You value culinary heritage, wood-fired standards, and simplicity. "
    "Speak with dramatic Italian passion, using phrases like 'Mamma mia!' and 'Sacrilege!'. "
    "Keep your response to 2-3 sentences. Do not use markdown headers."
)

ruben_sys = (
    "You are Ruben, an adventurous, rule-breaking modern foodie. "
    "You love sweet-and-savory flavor combinations and believe culinary evolution is a beautiful thing. "
    "You speak with high energy, enthusiasm, and references to caramelization, contrast, and culinary freedom. "
    "Keep your response to 2-3 sentences. Do not use markdown headers."
)

greg_sys = (
    "You are Greg, a highly sophisticated, slightly pretentious, and philosophical food critic. "
    "You analyze debates using gastronomy theory, flavor chemistry, and historical contexts. "
    "Speak in overly elegant, intellectual, and verbose sentences. "
    "Your goal in rounds 1 and 2 is to remain neutral and analyze their arguments. "
    "In round 3, you must make a final, definitive decision and side with either Ann or Ruben. "
    "Keep your response to 2-3 sentences. Do not use markdown headers."
)

# 2. General prompt runner using your free Gemini instance
def ask_panelist(system_instruction, transcript, role_name):
    prompt = f"{transcript}\n{role_name}, it is your turn to speak. Address the panel and build on the debate."
    response = client.models.generate_content(
        # model="gemini-3-flash-preview",
        model="gemini-2.5-flash",
        config={"system_instruction": system_instruction, "temperature": 0.85},
        contents=prompt
    )
    return response.text.strip()

# 3. Initialize history
transcript = f"The panel topic is: '{DEBATE_TOPIC}'\n\n"

display(Markdown(f"### 🎙️ The Great Pizza Panel: Three-Way Debate\n**Topic:** *{DEBATE_TOPIC}*\n---"))

# 4. Start the Debate Loop
for round_num in range(1, TURNS + 1):
    display(Markdown(f"#### 🍕 Round {round_num}"))
    
    # 👵🇮🇹 Ann's Turn
    ann_response = ask_panelist(ann_sys, transcript, "Ann")
    display(Markdown(f"👵🔴 **Ann**: *\"{ann_response}\"*"))
    transcript += f"Ann: {ann_response}\n\n"
    speak(ann_response, "Ann")

    
    # 🍍🍕 Ruben's Turn
    ruben_response = ask_panelist(ruben_sys, transcript, "Ruben")
    display(Markdown(f"🍍🟡 **Ruben**: *\"{ruben_response}\"*"))
    transcript += f"Ruben: {ruben_response}\n\n"
    speak(ruben_response, "Ruben")
    
    # 🎩🍷 Greg's Turn (Special prompt additions for his final verdict in Round 3)
    if round_num == TURNS:
        current_greg_sys = greg_sys + " This is the final round. You MUST definitively announce your winner now."
    else:
        current_greg_sys = greg_sys
        
    greg_response = ask_panelist(current_greg_sys, transcript, "Greg")
    display(Markdown(f"🎩🟣 **Greg**: *\"{greg_response}\"*"))
    transcript += f"Greg: {greg_response}\n\n"
    speak(greg_response, "Greg")


display(Markdown("--- \n### 🏁 The Panel has adjourned! Did Greg make the right choice?"))
```
